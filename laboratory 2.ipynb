```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("../input/test-imgs/amir temur.jpg")
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (12,12))
fig, ax = plt.subplots(figsize=(10, 10))

ax.imshow(img)
ax.axis('off')
plt.show()
```
